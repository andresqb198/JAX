```python
import matplotlib.pyplot as plt
import numpy as np

x_np = np.linspace(0,10,1000)
y_np = 2*np.sin(x_np)*np.cos(x_np)
plt.plot(x_np, y_np)
```

```python
import jax.numpy as jnp

x_jnp = jnp.linspace(0,10,100)
y_jnp = 2 * jnp.sin(x_jnp) * jnp.cos(x_jnp)
plt.plot(x_jnp,y_jnp)
```
